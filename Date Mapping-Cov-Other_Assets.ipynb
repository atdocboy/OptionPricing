```python
yield03=pd.read_csv(filepath + "Data of India Bond Yield  - 3 Month.csv")
```

```python
features['date'] = pd.to_datetime(features['date'])
features['date'] = features['date'].dt.date 
features = features.sort_values(by='date')
features.reset_index(inplace = True, drop = True)
# features.drop_duplicates(subset=['date'],inplace=True)
# features.reset_index(inplace = True, drop = True)
features
```

```python
yield03['Date'] = pd.to_datetime(yield03['Date'])
yield03['Date'] = yield03['Date'].dt.date 
yield03 = yield03.sort_values(by='Date')
yield03.reset_index(inplace = True, drop = True)
```

```python
def lookup(date,df):
    OG=df
    df=df.loc[df['Date']==date]
    df.reset_index(inplace = True, drop = True)
    while(df.empty):
        df=OG
#         print("Old:{}".format(date))
        date = date -  pd.to_timedelta(1, unit='d')
#         print("New:{}".format(date))
        df = df.loc[df['Date'] == date]
#         print(df)
    df.reset_index(inplace = True, drop = True)    
    return df['Price'][0]
#     print(df['Price'])
```

```python
cvList=[]
cov_list=['close_lr','open_lr','high_lr','low_lr']
for cl1 in range(len(cov_list)):
    for cl2 in range(cl1+1,len(cov_list)):
        cvl=cov_list[cl1]+"_"+cov_list[cl2]+"_cov"
        cvList.append(cvl)
# print(cvList)
col_name = ['date'] + ['close_mean','close_sd','open_mean','open_sd','high_mean','high_sd','low_mean','low_sd'] + cvList + ['change_oi','avg_atmError','ts_atr','prevOp'] + ['ttm', 'atmError','strike','spot','opPrice','percentage']
```

```python
featuresNew=pd.DataFrame(columns=col_name)
count=0

for _,row in features.iterrows():
    row['yield03']=lookup(row['date'],yield03)
    featuresNew=featuresNew.append(row)
```
